```python
import pandas as pd
import mplfinance as mpf
import investpy
import matplotlib
import datetime

```

```python
start_date = datetime.date(2021, 1, 1)
end_date = datetime.date(2021, 7, 31)
data = investpy.get_crypto_historical_data(crypto='bitcoin',
                                               from_date=str(start_date.strftime('%d/%m/%Y')),
                                               to_date=str(end_date.strftime('%d/%m/%Y')))
```

```python
data.info()
data
```

```python
mpf.plot(data)
```

```python
mpf.plot(data['2021-03':'2021-04'],mav =(20),type='candle',volume = True,tight_layout = True)
```

**Candlestick plot for march 2021**

```python
mpf.plot(data['2021-01':'2021-07']
         ,type='candle',
         mav = (2),
         volume = True,
         tight_layout = True,
         style = 'yahoo')
```
